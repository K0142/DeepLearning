```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
# 최소 제곱법 구현 

mx = np.mean(x)
my = np.mean(y)

divisor = sum([(i-mx)**2 for i in x])

def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d

dividend = top(x, mx, y, my)

a = dividend / divisor
b = my - (mx*a)

print("기울기 a = ", a)
print("y 절편 b = ", b)
```

```python
# 평균 제곱 오차(MSE) 구현

fake_a = 3
fake_b = 76

def predict(x):
    return fake_a*x + fake_b

predict_result = []

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부 시간 = %.f, 실제 점수 = %.f, 예측 점수 = %.f" % (x[i], y[i], predict(x[i])))
    
n = len(x)
def mse(y, pred_y):
    return (1/n)*sum((y-pred_y)**2)

print()
print("평균 제곱 오차: " + str(mse(y, predict_result)))
```

```python
# 선형 회귀 모델 실습

a = 0
b = 0
lr = 0.03
epochs = 2001
n = len(x)

for i in range(epochs):
    pred_y = a*x + b
    error = y - pred_y
    
    diff_a = (2/n) * sum(-x*(error))
    diff_b = (2/n) * sum(-(error))
    
    a = a - lr * diff_a
    b = b - lr * diff_b
    
    if i%100 == 0:
        print("epochs = %.f, 기울기 = %.04f, 절편 = %.04f" % (i, a, b))
    
pred_y = a*x + b

plt.scatter(x, y)
plt.plot(x, pred_y, 'r')
plt.show()
```

```python
# 다중 선형 회귀 모델 구현

x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 4])
y = np.array([81, 93, 91, 97])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y);
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0
lr = 0.01
epochs = 2001
n = len(x1)

for i in range(epochs):
    pred_y = a1*x1 + a2*x2 + b
    error = y - pred_y

    diff_a1 = (2/n) * sum(-x1 * (error))
    diff_a2 = (2/n) * sum(-x2 * (error))
    diff_b = (2/n) * sum( - (error))

    a1 = a1 - lr * diff_a1
    a2 = a2 - lr * diff_a2
    b = b - lr * diff_b
    
    if i%100 == 0:
        print("epochs = %.f, 기울기1 = %.04f, 기울기2 = %.04f, 절편 = %.04f" % (i, a1, a2, b))

print()
print("실제 점수 = ", y)
print("예측 점수 = ", pred_y)
```

```python
# Tensorflow를 이용한 선형 회귀 구현

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

model = Sequential()

model.add(Dense(1, input_dim = 1, activation = 'linear'))
model.compile(optimizer = 'sgd', loss = 'mse')
model.fit(x, y, epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

hour = 7
prediction = model.predict([hour])

print()
print("%.f시간을 공부할 경우 예상 점수는 %.02f점입니다." % (hour, prediction))
```

```python
# Tensorflow를 이용한 다중 선형 회귀 구현

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])

model = Sequential()

model.add(Dense(1, input_dim = 2, activation = 'linear'))
model.compile(optimizer = 'sgd', loss = 'mse')
model.fit(x, y, epochs=2000)

hour = 7
privation_class = 4
prediction = model.predict([[hour, privation_class]])

print()
print("%.f시간을 공부하고 %.f시간 과외를 받을 경우 예상 점수는 %.02f점입니다." % (hour, privation_class, prediction))
```

```python
# Tensorflow를 이용한 로지스틱 회귀 구현

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()
model.add(Dense(1, input_dim = 1, activation = 'sigmoid'))
model.compile(optimizer = 'sgd', loss = 'binary_crossentropy')
model.fit(x, y, epochs = 5000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%입니다." % (hour, prediction*100))
```
